Import all required modules

```python
from Data_modules import ModelClassifier, ModelMetrics
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

Initialize helpers and loading Wine data.

```python
trainer = ModelClassifier()
metrics = ModelMetrics()

wine_data = load_wine()
wine_df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
wine_df['wine_class'] = pd.Categorical.from_codes(wine_data.target, [0,1,2])
```

Check loaded data.

```python
wine_df
```

Prepeare data for models.

```python
X = wine_df.drop('wine_class', axis=1)
y = wine_df['wine_class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

Run all models on prepeared data. Result contains predicter values.

```python
result = trainer.get_all_models(X_train, X_test, y_train)
```

Run prediction metrics to check all models accuracy, classification report and confusion matrix.

```python
metrics_result = metrics.get_all_models_metrics(y_test, result)
```

Check all models accuracy together to select best model.

```python
metrics.get_all_models_accuracy()
```
