Load all neaded modules.

```python
from Data_modules import DataLoader, Visualizer, ModelRegressor, ModelMetrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
```

Load insurance data.

```python
datahandler = DataLoader()
trainer = ModelRegressor()
metrics = ModelMetrics()
enc = OrdinalEncoder()

insurance_df = datahandler.load_data('Data/insurance.csv')

insurance_df
```

Data preprocessing.

```python
enc.fit(insurance_df[['sex', 'smoker', 'region']])


insurance_df[['sex', 'smoker', 'region']] = enc.transform(insurance_df[['sex', 'smoker', 'region']])
insurance_df
```

Preparing train and test data for models.

```python
X = insurance_df.drop('charges', axis=1)
y = insurance_df['charges']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
result = trainer.get_all_models(X_train, X_test, y_train)
```

Evaluating models results.

```python
metrics_result = metrics.get_all_models_metrics(y_test, result)
```

Checking R2 metric.

```python
metrics.get_all_models_r2()
```

According to this results - Gradient Boosting models works best.
